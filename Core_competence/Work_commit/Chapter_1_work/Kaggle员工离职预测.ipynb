```python
# 删除无意义的变量，并检查剩余自字是否含有缺失值
#train.drop(['user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)
#第二种字段筛选
#train.drop(['Education','Gender','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','RelationshipSatisfaction','YearsSinceLastPromotion','user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)
#第三种字段筛选
train.drop(['Gender','MonthlyRate','user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)

#train.isnull().sum(axis = 0)
```

```python
# 删除无意义的变量，并检查剩余自字是否含有缺失值
#test.drop(['user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)
#第二种字段筛选
#test.drop(['Education','Gender','MonthlyRate','NumCompaniesWorked','PercentSalaryHike','RelationshipSatisfaction','YearsSinceLastPromotion','user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)
#第三种字段筛选
test.drop(['Gender','MonthlyRate','user_id','EmployeeCount','EmployeeNumber','HourlyRate','Over18','PerformanceRating','StandardHours'], axis = 1, inplace = True)

#test.isnull().sum(axis = 0)
```

```python
# 将数值型的字段转换为类别型，否则无法对其哑变量处理
train.BusinessTravel = train.BusinessTravel.astype('category')
train.Attrition = train.Attrition.astype('category')
train.Department = train.Department.astype('category')
train.EducationField = train.EducationField.astype('category')
#train.Gender = train.Gender.astype('category')
train.JobRole = train.JobRole.astype('category')
train.MaritalStatus = train.MaritalStatus.astype('category')
train.OverTime = train.OverTime.astype('category')
```

```python
# 将数值型的字段转换为类别型，否则无法对其哑变量处理
test.BusinessTravel = test.BusinessTravel.astype('category')
test.Department = test.Department.astype('category')
test.EducationField = test.EducationField.astype('category')
#test.Gender = test.Gender.astype('category')
test.JobRole = test.JobRole.astype('category')
test.MaritalStatus = test.MaritalStatus.astype('category')
test.OverTime = test.OverTime.astype('category')
```

```python
#哑变量处理
#dummy = pd.get_dummies(train[['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']])
dummy = pd.get_dummies(train[['Attrition','BusinessTravel','Department','EducationField','JobRole','MaritalStatus','OverTime']])
```

```python
#哑变量处理
#dummy_test = pd.get_dummies(test[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']])
dummy_test = pd.get_dummies(test[['BusinessTravel','Department','EducationField','JobRole','MaritalStatus','OverTime']])
```

```python
# 水平合并train数据集和哑变量的数据集
train = pd.concat([train,dummy], axis = 1)
```

```python
# 水平合并train数据集和哑变量的数据集
test = pd.concat([test,dummy_test], axis = 1)
```

```python
# 删除原始的字段变量
#train.drop(['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'], inplace=True, axis = 1)
train.drop(['Attrition','BusinessTravel','Department','EducationField','JobRole','MaritalStatus','OverTime'], inplace=True, axis = 1)
#train.head()
```

```python
# 删除原始的字段变量
#test.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'], inplace=True, axis = 1)
test.drop(['BusinessTravel','Department','EducationField','JobRole','MaritalStatus','OverTime'], inplace=True, axis = 1)
#test.head()
```

```python
predictor = train.drop(['Attrition_No','Attrition_Yes'], inplace=False, axis = 1)
```

```python
# 导入第三方包
from sklearn import model_selection
# 将数据集拆分为训练集和测试集，且测试集的比例为25%
#X_train, X_test, y_train, y_test = model_selection.train_test_split(predictor, train.Attrition_Yes, 
                                                                    #test_size = 0.30, random_state = 1234)
```

```python
X_train = predictor
y_train = train.Attrition_Yes
X_test = test
```

```python
# 导入第三方模块
from sklearn.model_selection import GridSearchCV
from sklearn import tree
# 预设各参数的不同选项值
max_depth = [2,3,4,5,6]
min_samples_split = [2,4,6,8]
min_samples_leaf = [2,4,8,10,12]
# 将各参数值以字典形式组织起来
parameters = {'max_depth':max_depth, 'min_samples_split':min_samples_split, 'min_samples_leaf':min_samples_leaf}
# 网格搜索法，测试不同的参数值
grid_dtcateg = GridSearchCV(estimator = tree.DecisionTreeClassifier(), param_grid = parameters, cv=10)
# 模型拟合
grid_dtcateg.fit(X_train, y_train)
# 返回最佳组合的参数值
grid_dtcateg.best_params_
```

```python
# 导入第三方模块
from sklearn import metrics
# 构建分类决策树
CART_Class = tree.DecisionTreeClassifier(max_depth=6, min_samples_leaf = 10, min_samples_split=2)
# 模型拟合
decision_tree = CART_Class.fit(X_train, y_train)
t = CART_Class.predict_proba(X_test)
```
