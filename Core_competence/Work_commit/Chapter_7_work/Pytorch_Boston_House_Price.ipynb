```python
from sklearn.datasets import load_boston
data = load_boston()
```

```python
x = data['data']
y = data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
#转成list格式
y = y.reshape(-1,1)
```

```python
print(y)
```

```python
#数据规范化
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()
x = ss.fit_transform(x)
```

```python
#将numpy数据放入到torch种
import torch
from torch import nn
x = torch.from_numpy(x).type(torch.FloatTensor)
y = torch.from_numpy(y).type(torch.FloatTensor)
#数据集切分
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(x,y,test_size = 0.2)
```

```python
#构造网络
model = nn.Sequential(
        nn.Linear(13,18),#
        nn.ReLU(),
        nn.Linear(18,1)
        
)
#定义优化器和损失函数
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(),lr = 0.01)
```

```python
print(train_X.shape)
```

```python
print(model(train_X).shape)
```

```python
print(train_y.shape)
```

```python
print(x.shape)
```

```python
#训练
max_epoch = 1000
iter_loss = []
for i in range(max_epoch):
    #前向传播
    y_pred = model(train_X)
    #计算损失
    loss = criterion(y_pred,train_y)
    print(i,loss.item())
    iter_loss.append(loss.item())
    #梯度清零（针对Pytorch）
    optimizer.zero_grad()
    #反向传播
    loss.backward()
    #梯度调整
    optimizer.step()
    
```

```python
#测试
output = model(test_X)
#将数据从Pytorch放入到numpy
predict_list = output.detach().numpy()
print(predict_list)
```

```python
#绘制不同的iteration_loss
import numpy as np
x = np.arange(max_epoch)
y = np.array(iter_loss)
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title('Loss value in all iterations')
plt.xlabel('Iterations')
plt.ylabel('Mean Loss Value')
plt.show()
```

```python
#查看真实值与预测值的散点图
x = np.arange(test_X.shape[0])
y1 = np.array(predict_list)#预测值
y2 = np.array(test_y)#实际值
line1 = plt.scatter(x,y1,c='red')
line2 = plt.scatter(x,y2,c='yellow')
plt.legend([line1,line2],['predict','real'])
plt.title('Prediction vs Real')
plt.ylabel('Boston House Price')
plt.show()
```
